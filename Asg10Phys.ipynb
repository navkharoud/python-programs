```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
```

```python
# using built in data sets
print(sns.get_dataset_names())
```

```python
planets_df = sns.load_dataset('planets')
planets_df
```

```python
sns.histplot(planets_df['mass'], kde = True)
```

```python
# plotting a regression line
# on x axis : distance of the planet and on y-axis: mass of the planet
sns.jointplot(x='distance', y='mass', data = planets_df, kind = 'reg')
```

```python
# plotting a kde
# on x axis : distance of the planet and on y-axis: mass of the planet
sns.jointplot(x='distance', y='mass', data = planets_df, kind = 'kde')
```

```python
# kde plot for mass
sns.kdeplot(planets_df['mass'])
```

```python
# plotting a hexagon 
# on x axis : distance of the planet and on y-axis: mass of the planet
sns.jointplot(x='distance', y='mass', data = planets_df, kind = 'hex')
```

```python
# pair plot 
sns.pairplot(planets_df)
```
